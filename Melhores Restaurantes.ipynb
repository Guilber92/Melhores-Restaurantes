```python
import matplotlib.pyplot as plt

# Contar a quantidade de restaurantes por país
contagem_paises = df['country'].value_counts()

# Criar o gráfico de barras
plt.figure(figsize=(14, 10))
plt.bar(contagem_paises.index, contagem_paises.values, color='orange')
plt.title('location')
plt.xlabel('country')
plt.ylabel('rank')
plt.xticks(rotation=90)  # Rotacionar os rótulos dos países para melhor visualização
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Carregar o DataFrame
df = pd.read_csv('WorldsBestRestaurants.csv')

# Agrupar os dados por país e contar a frequência
counts = df['country'].value_counts()

# Selecionar os 5 primeiros países (os com maior frequência)
top5_countries = counts.head(5)

# Criar o gráfico de pizza
plt.pie(top5_countries, labels=top5_countries.index, autopct='%1.1f%%')  # autopct mostra a porcentagem em cada fatia
plt.title('5 Países com Mais Restaurantes')
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Carregar os dados
df = pd.read_csv('WorldsBestRestaurants.csv')

# Filtrar para os Estados Unidos (ajuste o nome da coluna 'country' se necessário)
usa_restaurants = df[df['country'] == 'United States']

# Agrupar por cidade e contar a frequência
counts_usa = usa_restaurants.groupby('location').size()

# Criar o mapa de calor
plt.figure(figsize=(12, 8))
sns.heatmap(counts_usa.to_frame(), annot=True, fmt='d', cmap='YlOrRd')
plt.title('Concentração de Restaurantes nos Estados Unidos por Cidade')
plt.xlabel('Cidade')
plt.ylabel('Número de Restaurantes')
plt.show()
```
